## cape Outliers

```python
cols = data.select_dtypes(include=np.number).columns.tolist()
num_subplots = len(cols)
num_rows = (num_subplots + 1) // 2  #extra one for odd no. of plots
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))

axes = axes.flatten()

for colIndex in range(num_subplots):
    axes[colIndex].boxplot(data[cols[colIndex]])
    axes[colIndex].set_xlabel(cols[colIndex])

plt.show()
```

```python
for col in cols:
  ll,ul=data[col].quantile([.01,.99])
  data[col]=np.where(data[col]<ll,ll,data[col])
  data[col]=np.where(data[col]>ul,ul,data[col])
```

since for most of the features 99th percentile falls beyond whiskers, they are still outliers but they make sense

## Encode categorical values to numerical

state and season to dummies/OneHotEncoder <br>
since crop is predictor it will be label encoding

```python
data=pd.get_dummies(data,columns=['State','Season'])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['Crop'] = label_encoder.fit_transform(data['Crop'])
```

# Traning and testing Data

```python
features = data.drop('Crop', axis=1)
target = data['Crop']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
# Create and fit the decision tree model for classification
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions on the training set
train_predictions = model.predict(X_train)
# Make predictions on the test set
test_predictions = model.predict(X_test)
```

```python
# Accuracy
train_accuracy = accuracy_score(y_train, train_predictions) * 100
test_accuracy = accuracy_score(y_test, test_predictions) * 100
# Display the results
print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

```python
# Precision
train_precision = precision_score(y_train, train_predictions, average='weighted') * 100
test_precision = precision_score(y_test, test_predictions, average='weighted') * 100

# Display the results
print("\nTraining Precision:", train_precision)
print("Testing Precision:", test_precision)
```

```python
# Recall
train_recall = recall_score(y_train, train_predictions, average='weighted') * 100
test_recall = recall_score(y_test, test_predictions, average='weighted') * 100

# Display the results
print("\nTraining Recall:", train_recall)
print("Testing Recall:", test_recall)
```

```python
# F1 Score
train_f1 = f1_score(y_train, train_predictions, average='weighted') * 100
test_f1 = f1_score(y_test, test_predictions, average='weighted') * 100

# Display the results
print("\nTraining F1 Score:", train_f1)
print("Testing F1 Score:", test_f1)
```

```python
# Confusion Matrix
conf_matrix_train = confusion_matrix(y_train, train_predictions)
conf_matrix_test = confusion_matrix(y_test, test_predictions)

# Display the results
print("\nConfusion Matrix (Training):\n", conf_matrix_train)
print("\nConfusion Matrix (Testing):\n", conf_matrix_test)
```

```python
result_train=pd.DataFrame(columns = ['Algorithm','train_accuracy','train_precision','train_recall','train_f1'])
result_test=pd.DataFrame(columns = ['Algorithm','test_accuracy','test_precision','test_recall','test_f1'])
result_train.loc[0] = ['Desicion Tree Classifier',train_accuracy, train_precision, train_recall, train_f1]
result_test.loc[0] = ['Desicion Tree Classifier',test_accuracy, test_precision, test_recall, test_f1]
```
